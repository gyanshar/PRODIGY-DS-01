#### Data Pre-Processing:

Importing the libraries:

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns 
sns.set_theme()

pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
```

Importing the dataset:

```python
df1 = pd.read_csv('API_SP.POP.TOTL_DS2_en_csv_v2_2544372.csv')
df2 = pd.read_csv('Metadata_Country_API_SP.POP.TOTL_DS2_en_csv_v2_2544372.csv')
```

Merging both the datasets:

```python
df = pd.merge(df1, df2, on='Country Code')
df.head()
```

#### Data Cleaning:

```python
df.columns
```

```python
df.drop(['Unnamed: 68', 'Indicator Name', 'Indicator Code', 'Unnamed: 5', 'TableName', 'SpecialNotes'], axis=1, inplace=True)
```

```python
df.info()
```

```python
df.shape
```

Checking for duplicates:

```python
df.duplicated().sum()
```

Checking for null values:

```python
df.isnull().sum()
```

```python
df[df['1970'].isnull()==True]
```

```python
df[(df['Country Name'] == 'West Bank and Gaza')]
```

```python
df.drop(df[df['Country Name'] == 'West Bank and Gaza'].index, axis=0, inplace=True)
```

Cleaning the Country Names Column:

```python
df.drop(df[(df['Country Name'].str.lower().str.contains('income'))].index, axis=0, inplace=True)
df.drop(df[(df['Country Name'].str.lower().str.contains('ida '))].index, axis=0, inplace=True)
df.drop(df[(df['Country Name'].str.lower().str.contains('members'))].index, axis=0, inplace=True)
df.drop(df[(df['Country Name'].str.lower().str.contains('developed'))].index, axis=0, inplace=True)
df.drop(df[(df['Country Name'].str.lower().str.contains('only'))].index, axis=0, inplace=True)
df.drop(df[(df['Country Name'].str.lower().str.contains('world'))].index, axis=0, inplace=True)
df.drop(df[(df['Country Name'].str.lower().str.contains('dividend'))].index, axis=0, inplace=True)
df.drop(df[(df['Country Name'].str.lower().str.contains('situations'))].index, axis=0, inplace=True)
df.drop(df[(df['Country Name'].str.lower().str.contains('area'))].index, axis=0, inplace=True)
df.drop(df[(df['Country Name'].str.lower().str.contains('&'))].index, axis=0, inplace=True)
df.drop(df[(df['Country Name'].str.lower().str.contains('indebted'))].index, axis=0, inplace=True)
df.drop(df[(df['Country Name'].str.lower().str.contains('sub'))].index, axis=0, inplace=True)
df.drop(df[(df['Country Name'].str.lower().str.contains('western'))].index, axis=0, inplace=True)
df.drop(df[(df['Country Name'].str.lower().str.contains('eastern'))].index, axis=0, inplace=True)
df.drop(df[(df['Country Name'].str.lower().str.contains('european'))].index, axis=0, inplace=True)
df.drop(df[(df['Country Name'].str.lower().str.contains('south asia'))].index, axis=0, inplace=True)
```
